**2.1 Установить характер разрыва функции в точке x0:**

$$1)f(x)=\frac{x^2−16}{x+4}, x0 = −4;$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math


x = np.linspace(-40, 40, 10000)
y = (x**2 - 16)/(x + 4)
plt.figure(1)
plt.title('График функции y = (x**2 - 16)/(x + 4)')
plt.ylabel('Ось y')
plt.xlabel('Ось x')
plt.grid()
plt.axis([-20, 20, -20, 20]) 
plt.plot(x, y)
plt.show()
```

Установить характер разрыва функции в точке x0 не возможно, так как функция неразрывна.

**2.2 Установить характер разрыва функции в точке x0:**

**3.1 Исследовать на непрерывность функцию f(x) в точке x0:**


$$1)f(x)=arctg\,\frac{2}{x-1},\,\:x_{(0)}=1;$$

**3.2 Исследовать на непрерывность функцию f(x) в точке x0:**
